```python
#ITU-R P.837
#Calculating P0 annual
##converted to function

def P0(Lat,Lon):

  import numpy as np
  month=[[1,31],[2,28.25],[3,31],[4,30],[5,31],[6,30],[7,31],[8,31],[9,30],[10,31],[11,30],[12,31]] #list in format[ii,Ni]
  Months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
  N=[31,28.25,31,30,31,30,31,31,30,31,30,31]
  T=[287.153563, 290.519574, 295.963952,302.354184, 306.048907, 306.507869, 303.636607, 302.656573, 301.789416, 298.546718, 293.158063, 288.463691 ]  #Tii values for sample, to be taken from local data                    #TBE
  MT=[17.627556, 14.730875, 10.138234,6.914227, 8.625949, 36.161384, 188.164508, 183.523770, 109.466133, 18.547533, 2.657261, 3.967131] #MTii values for sample, to be taken from local data                               #TBE
  R=[]
  P=[]

  #ppercent=float(input('p:Desired annual probability of exceedance'))

  #Step 4
  for i in range(0,12):
    T[i]=T[i]-273.15 #Tii list converted from k to C

  #calculating rii values (Step5)
  for j in T:     #j is tii in the formula
    if j>=0:
      r=0.5874*np.exp(0.0883*j)
      R.append(r)
    elif j<0:
      r=0.5874
      R.append(r) #rii values list

  #calculating Pii values (Step6A)
  for k in range(0,12):
    p=100*MT[k]/(24*N[k]*R[k]) 
    P.append(p) #adding Pii values in list of p

  #Step 6B
  for l in range(0,12):
    if P[l]>70:
      P[l]=70
      R[l]=5*MT[l]/(84*N[l])

  #Step 7 Calculating P annual
  sum=0
  for m in range(0,12):
    x=P[m]*N[m] #x denotes the product of Nii and P0ii
    sum=sum+x
  Pannual=sum/365.25
  print(Pannual)
  return(Pannual)
  #if p>Pannual:
  #  Rp=0
  #elif p<=Pannual:
P0(28.6517178,77.2219388)
```

```python
######converted to function

def kalpha(f,Lat,Lon,LonS):
  import math
  f=15
  phid=Lat     #latitude of the earth station (degrees) 
  phi=math.radians(phid) 
  LonS=74 ###TBE satellite position in degrees    (74 degree) 
  delta=(math.radians(LonS))-(math.radians(Lon))         
  numcos=(math.cos(delta))*(math.cos(phi))
  theta=math.atan((numcos-0.1512)/math.sqrt(1-numcos**2))
  thetad=math.degrees(theta)
  print(thetad)
  towd=45
  tow=math.radians(towd)


####ITU-R P.838

#aj,bj,cj, ck and mk values for kH
  akh=[-5.33980,-0.35351,-0.23789,-0.94158]
  bkh=[-0.10008,1.26970,0.86036,0.64552]
  ckh=[1.13098,0.45400,0.15354,0.16817]
  mkh=-0.18961
  cnstckh=0.71147

#kH calculation
  sumkh=0
  for i in range(0,4):
    sumtempkh=0
    powertempkh=-(((math.log(f,10)-bkh[i])/ckh[i])**2)
    sumtempkh=akh[i]*math.exp(powertempkh)
    sumkh=sumkh+sumtempkh

  logkh=sumkh+(mkh*math.log(f,10))+cnstckh
  kh=math.pow(10,logkh)
  print('kh',kh)

###################

#aj,bj,cj, ck and mk values for kV
  akv=[-3.80595,-3.44965,-0.39902,0.50167]
  bkv=[0.56934,-0.22911,0.73042,1.07319]
  ckv=[0.81061,0.51059,0.11899,0.27195]
  mkv=-0.16398
  cnstckv=0.63297

#kV calculation
  sumkv=0
  for i in range(0,4):
    sumtempkv=0
    powertempkv=-(((math.log(f,10)-bkv[i])/ckv[i])**2)
    sumtempkv=akv[i]*math.exp(powertempkv)
    sumkv=sumkv+sumtempkv

  logkv=sumkv+(mkv*math.log(f,10))+cnstckv
  kv=math.pow(10,logkv)
  print('kv',kv)

########################

#aj,bj,cj, c alpha and m alpha values for alpha H
  aah=[-0.14318,0.29591,0.32177,-5.37610,16.1721]
  bah=[1.82442,0.77564,0.63773,-0.96230,-3.29980]
  cah=[-0.55187,0.19822,0.13164,1.47828,3.43990]
  mah=0.67849
  cnstcah=-1.95537

#alpha H calculation
  sumah=0
  sumtempah=0
  for j in range(0,5):
  #sumtempah=0
    powertempah=-(((math.log(f,10)-bah[j])/cah[j])**2)
    sumtempah=aah[j]*math.exp(powertempah)
    sumah=sumah+sumtempah

  ah=sumah+(mah*math.log(f,10))+cnstcah
  print('ah',ah)


########################


#aj,bj,cj, c alpha and m alpha values for alpha V
  aav=[-0.07771,0.56727,-0.20238,-48.2991,48.5833]
  bav=[2.33840,0.95545,1.14520,0.791669,0.791459]
  cav=[-0.76284,0.54039,0.26809,0.116226,0.116479]
  mav=-0.053739
  cnstcav=0.83433


#alpha V calculation
  sumav=0
  sumtempav=0
  for j in range(0,5):
   #sumtempav=0
    powertempav=-1*(((math.log(f,10)-bav[j])/cav[j])**2)
    sumtempav=aav[j]*math.exp(powertempav)
    sumav=sumav+sumtempav

  av=sumav+(mav*math.log(f,10))+cnstcav
  print('av',av)

#################


#k and alpha values final caluclation

  c1=math.cos(theta)**2
  c2=math.cos(2*tow)

  k=(kh+kv+((kh-kv)*c1*c2))/2
  alpha=((kh*ah)+(kv*av)+(((kh*ah)-(kv*av))*c1*c2))/(2*k)
  print('k:',k)
  print('alpha',alpha)

kalpha(15,28.65,77.22,74)

```

#Step A2


```python
import math

#all constants used

h0=4.9123#using BLI in matlab
hr=h0+0.36         #######ITU-R P.839
hs=0.225    ######calculated using BLI as stated in ITU-R P.1511 #require TOPO0DOT5 file
rs=hr-hs
Re=8500                       #radius of earth in km
RO=35786 #km
R=61.8779 #R0.01 value
#R=r001    
AiResearchpaper=[]                                          



#ITU-R P.618
#Calculating Ai and Pi pairs


##Pi values

PiList=[0.01,0.02,0.03,0.05,0.1,0.2,0.3,0.5,1,2,3,5,10]
AiList=[]
PiList = [ elem for elem in PiList if elem<=Pannual]
print('Pi List:',PiList)

Plen=len(PiList)
for i in range(0,Plen):
  Pi=PiList[i]
###Ai values calculation for different Pi
#For Pi=0.01%

  if Pi==0.01:
    if rs<=0:
      A=0
  #Step 2
    else:
      if thetad>=5:
        Ls=(rs)/math.sin(theta)
      elif thetad<5:
        Ls=2*rs/(math.sqrt((math.sin(theta)**2)+(2*rs/Re))+math.sin(theta))

      if Ls<0:
        Ls=0
      else:
        Ls=Ls
    #Step 3
      Lg=Ls*math.cos(theta)

  #Step 4

    if R==0:
      A=0
  #Step 5:

    else:
      yr=k*pow(R,alpha) #gamma     
      print('gamma:',yr)                                    
  #Step 6
    r=1/(1+(0.78*(math.sqrt(Lg*yr/f)))-(0.38*(1-(math.exp(-2*Lg)))))  #horizontal reduction factor

  #Step 7:
    zeta=math.degrees(math.atan(rs/(Lg*r)))
  ##Lr calculation
    if zeta>thetad:
      Lr=(Lg*r)/math.cos(theta)
    else:
      Lr=rs/math.sin(theta)

##Chi calculation
    if abs(phid)<36:
      chi=36-abs(phid)
    else:
      chi=0

##v 0.01 calculation
    # v=1/(1+((math.sqrt(math.sin(theta)))*(31*(1-(math.exp((-thetad)/(1+chi)))*((math.sqrt(Lr*yr))/(f*f)))-0.45))) 
    v=1/(1+((math.sqrt(math.sin(theta)))*((31*(1-(math.exp(-1*(thetad/(1+chi)))))*((math.sqrt(Lr*yr))/(f*f)))-0.45))) ###############################

####  r: path reduction factor  #test
    rtest=1/((-0.3979/math.sin(theta))+(0.0021*R*rs)-0.0185+0.2337)
    A001test=yr*Ls*rtest
    print('A0.001',A001test)
    AiResearchpaper.append(A001test)

##Le calculation
    Le=Lr*v

###Attenuation calculation A0.01
    A=yr*Le                                         #Ai in dB
    AiList.append(A)

###For other values of Pi
  else:
  ### Beta values calculation
    if Pi>=1 or abs(phid)>=36:
      betaf=0
    elif Pi<1 and abs(phid)<36 and thetad>=25:
      betaf=-0.005*(abs(phid)-36)
    else:
      betaf=-0.005*(abs(phid)-36)+1.8-(4.25*math.sin(theta))
  
    Ap=A*pow((Pi/0.01),-(0.655+(0.033*math.log(Pi))-(0.045*math.log(A))-(betaf*(1-Pi)*math.sin(theta))))   #Ai in dB

    # Ai using reduction factor r
    Aptest=A001test*pow((Pi/0.01),-(0.655+(0.033*math.log(Pi))-(0.045*math.log(A))-(betaf*(1-Pi)*math.sin(theta))))   #Ai in dB
   
    AiList.append(Ap)
    AiResearchpaper.append(Aptest)
print('Ai List:',AiList)
print('\nAi using path reduction factor "r":',AiResearchpaper)


###List of Pi Ai pairs #nested list
PiAiList=[]
for i in range(0,Plen):
  templist=[]
  templist.append(PiList[i])
  templist.append(AiList[i])    #All Ai values in dB
  PiAiList.append(templist)

print(PiAiList)
print('Ls:', Ls, 'Lg:', Lg, 'Lr:', Lr, 'Le', Le)
  
```
